**Ejercicio 1:**

1.1 Piensa en 7 comidas diferente que te gusten. Implementa un programa que pregunte por las 7 comidas, las almacene en una lista y que posteriermente imprima cada una de ellas utilizando uno de los ciclos comentados en clase. 

```python
i=0
lista = list () #se crea la lista
while i<7: #se crea un ciclo while para imprimir 7 veces la pregunta de la comida
  comida = input ('¿Cuál es una de tus comidas favoritas?')
  lista.append(comida) #se agregan a la lista cada respuesta de la comida 
  i+=1 #se le agrega 1 a i para aumentar el ciclo
print("Las 7 comidas favoritas son:") 
for comida in lista:
  print(comida) #se imprime cada comida utilizando la función for
```
